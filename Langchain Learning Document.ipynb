```python
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import CharacterTextSplitter
from langchain import OpenAI
from langchain.document_loaders import DirectoryLoader
from langchain.chains import RetrievalQA

# Loading all files of type txt in the folder
loader = DirectoryLoader('C:/Users/jianbai/Downloads/Novels', glob='**/*.txt')
# Converting data into document objects, each document will be treated as a document
documents = loader.load()

# Initializing the loader
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
# Cutting the loaded document
split_docs = text_splitter.split_documents(documents)

# Initializing openai's embeddings object
embeddings = OpenAIEmbeddings()
# Calculating the embedding vector information from the openai's embeddings object and temporarily store the document into the Chroma vector database for subsequent matching queries.
docsearch = Chroma.from_documents(split_docs, embeddings)

# Creating Questioning Objects
qa = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=docsearch.as_retriever(), return_source_documents=True)
# Questioning
result = qa({"query": "Who is the main character of 雅量第六"})
print(result)
```

```python
import os

from langchain.document_loaders import YoutubeLoader
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import ChatVectorDBChain, ConversationalRetrievalChain

from langchain.chat_models import ChatOpenAI
from langchain.prompts.chat import (
  ChatPromptTemplate,
  SystemMessagePromptTemplate,
  HumanMessagePromptTemplate
)

# Loading youtube channel
loader = YoutubeLoader.from_youtube_url('https://www.youtube.com/watch?v=Dj60HHy-Kqk')
# Converting data to document
documents = loader.load()

# Initializing the text splitter
text_splitter = RecursiveCharacterTextSplitter(
  chunk_size=1000,
  chunk_overlap=20
)

# Splitting youtube files
documents = text_splitter.split_documents(documents)

# Initializing openai embeddings
embeddings = OpenAIEmbeddings()

# Storing data into vector storage
vector_store = Chroma.from_documents(documents, embeddings)
# Initializing the Retriever with Vector Storage
retriever = vector_store.as_retriever()

system_template = """
Use the following context to answer the user's question.
If you don't know the answer, say you don't, don't try to make it up. And answer in Chinese.
-----------
{question}
-----------
{chat_history}
"""

# Constructs the initial list of messages, which can be interpreted here as the messages parameter passed in by openai.
messages = [
  SystemMessagePromptTemplate.from_template(system_template),
  HumanMessagePromptTemplate.from_template('{question}')
]

# Initialize the prompt object
prompt = ChatPromptTemplate.from_messages(messages)


# Initializing the Q&A Chain
qa = ConversationalRetrievalChain.from_llm(ChatOpenAI(temperature=0.1,max_tokens=2048),retriever,condense_question_prompt=prompt)


chat_history = []
while True:
  question = input('Question：')
  # Start sending questions chat_history is a required parameter, used to store the conversation history.
  result = qa({'question': question, 'chat_history': chat_history})
  chat_history.append((question, result['answer']))
  print(result['answer'])
```
